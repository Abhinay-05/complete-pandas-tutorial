```python

bioSet['born_date'] = pd.to_datetime(bioSet['born_date'], errors='coerce')
bioSet['born_year'] = bioSet['born_date'].dt.year
# create new CSV without any added index
# bioSet.to_csv('./data/bioSet.csv', index=False)
bioSet.head()
```

```python
bioSet['height_category'] = bioSet['height_cm'].apply(
    lambda x: "Nan" if pd.isna(x) else("Short" if x<165 else("Average" if x<180 else "Tall"))
)
bioSet.sort_values('height_cm').tail()
bioSet.sort_values('height_cm').tail()
```

```python
def categorize_athelete(row):
    if row['height_cm'] == pd.isna(row['height_cm']) or row['weight_kg'] == pd.isna(row['weight_kg']):
        return "NaN"
    elif row['height_cm'] < 175 and row['weight_kg'] < 70:
        return "Light Weight"
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
        return "Middle Weight"
    else:
        return "Heavy Weight"
    
# bioSet['Category'] = bioSet.apply(categorize_athelete, axis=1)    
bioSet['height_cm'] == "NaN"
```
